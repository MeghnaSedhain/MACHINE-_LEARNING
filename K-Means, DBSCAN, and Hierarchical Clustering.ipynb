```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
```

```python
# Generate sample data with 3 clusters
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Visualize raw data
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], s=30)
plt.title("Original Scaled Data")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Visualize KMeans result
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='viridis')
plt.title("K-Means Clustering")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Visualize DBSCAN result
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_labels, cmap='rainbow')
plt.title("DBSCAN Clustering")
plt.show()
```

```python
agg = AgglomerativeClustering(n_clusters=3)
agg_labels = agg.fit_predict(X_scaled)

# Visualize Agglomerative Clustering result
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=agg_labels, cmap='plasma')
plt.title("Hierarchical Clustering (Agglomerative)")
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Create linkage matrix
linked = linkage(X_scaled, method='ward')

# Plot dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```
